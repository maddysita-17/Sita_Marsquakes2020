```python
from obspy.core.trace import Trace, Stats
from obspy.core.stream import Stream
from obspy import UTCDateTime
import obspy
from obspy.clients.fdsn import Client


client = Client("IRIS")

def waves_comp(start, end, adjtime):
    st = client.get_waveforms("II", "JTS", "10", "LH*", start-adjtime, end+adjtime, attach_response=True)
    #st.detrend(type='simple')
    return st

def waves_comp_mars(start, end, adjtime):
    st = client.get_waveforms('XB','ELYSE','02','B*', start-adjtime, end+adjtime, attach_response=True)
    #st.detrend(type='simple')
    return st
```

```python
arrival = UTCDateTime('2020-04-01T00:00:00')
begin = arrival             #p-wave arrival
end = arrival + 420        #s-wave arrival

st_idaho = waves_comp(begin, end, 800)
```

```python
st_idaho.plot();
```

```python
%matplotlib notebook
st_disp = st_idaho.copy()
st_disp.remove_response(output='DISP')
st_disp.plot()
```

```python
arrival = UTCDateTime('2019-10-26T02:15:50')
begin = arrival
end = arrival + 174

st_mars = waves_comp_mars(begin,end,800)
```

```python
%matplotlib notebook
st_disp = st_mars.copy()
st_disp.remove_response(output='DISP')
st_disp.plot()
```
