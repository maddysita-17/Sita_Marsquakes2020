```python
import matplotlib.pyplot as plt
import numpy as np
import obspy
import pandas as pd

from obspy.imaging.beachball import beachball
from obspy.imaging.beachball import beach
from obspy.imaging.source import plot_radiation_pattern
```

```python
def bbplot(file, event):
    faults = pd.read_csv(file, usecols=[0,1,2])

    faults_ls =[]

    for index, rows in faults.iterrows():
        my_list =[rows.Strike, rows.Dip, rows.Rake]
        faults_ls.append(my_list)

    n = 0
    for fault in faults_ls:
        if n == 0:
            fig = beachball(fault, alpha = 0.5, edgecolor = None)

        else:
            fig2 = beachball(fault, alpha=0.5, edgecolor = None, fig=fig)
        n += 1
        
    plt.savefig('beachball_' + event + '.png')
    return
```

```python
%matplotlib notebook
bbplot('uniquefaults_173a.csv', '173a')
```
