```python
import matplotlib.pyplot as plt
import numpy as np
from obspy.taup import TauPyModel
import pandas as pd
```

```python
# EH45: 235b @ 45km
radius = 3389.5
mars = TauPyModel(model="EH45Tcold")
mtimes = mars.get_travel_times(source_depth_in_km=45, distance_in_degree=27.5, phase_list=["P", "S"])
print(mars.model_name)
```

```python
%matplotlib notebook
arrivals = mars.get_ray_paths(
    source_depth_in_km=45, distance_in_degree=27.5, phase_list=["P", "S"])
ax = arrivals.plot_rays()
ax = arrivals.plot_rays(plot_type='cartesian')
```
